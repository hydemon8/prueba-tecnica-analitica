# Tabla de posiciones Mundial 1991

```python
matches_1991 = matches[matches['Year'] == 1991].copy()
```

```python
home_teams = matches_1991['home_team'].unique()
away_teams = matches_1991['away_team'].unique()
all_teams_1991 = pd.Series(list(set(home_teams) | set(away_teams))).sort_values().tolist()
```

```python
tabla_posiciones = pd.DataFrame({
    'Equipo': all_teams_1991,
    'PJ': 0,
    'PG': 0,
    'PE': 0,
    'PP': 0,
    'GF': 0,
    'GC': 0,
    'DG': 0,
    'JL': 0,
    'Puntos': 0
})
```

```python
print(matches_1991[['home_yellow_card_long', 'away_yellow_card_long', 
                     'home_red_card', 'away_red_card',
                     'home_yellow_red_card', 'away_yellow_red_card']].head(5))
```

```python
print("Amarillas-rojas en 1991:")
print(matches_1991[['home_yellow_red_card', 'away_yellow_red_card']].dropna())
```

```python
def contar_tarjetas(valor):
    if pd.isna(valor) or valor == '':
        return 0
    valor_limpio = valor.replace(' ,', ',').replace(', ', ',')
    return len(valor_limpio.split("', "))
```

```python
for idx, partido in matches_1991.iterrows():
    home = partido['home_team']
    away = partido['away_team']
    home_score = partido['home_score']
    away_score = partido['away_score']
    
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'PJ'] += 1
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'PJ'] += 1
    
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'GF'] += home_score
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'GC'] += away_score
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'GF'] += away_score
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'GC'] += home_score
    
    if home_score > away_score:  #gana local
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'PG'] += 1
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'Puntos'] += 3
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'PP'] += 1
    elif home_score < away_score:  #gana visitante
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'PG'] += 1
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'Puntos'] += 3
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'PP'] += 1
    else:  #empate
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'PE'] += 1
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'PE'] += 1
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'Puntos'] += 1
        tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'Puntos'] += 1
    
    #esta parte es juego limpio
    home_yellow = contar_tarjetas(partido['home_yellow_card_long'])
    away_yellow = contar_tarjetas(partido['away_yellow_card_long'])
    home_red = contar_tarjetas(partido['home_red_card'])
    away_red = contar_tarjetas(partido['away_red_card'])

    tabla_posiciones.loc[tabla_posiciones['Equipo'] == home, 'JL'] += (home_yellow * -1) + (home_red * -2)
    tabla_posiciones.loc[tabla_posiciones['Equipo'] == away, 'JL'] += (away_yellow * -1) + (away_red * -2)
```

```python
tabla_posiciones['DG'] = tabla_posiciones['GF'] - tabla_posiciones['GC']

tabla_posiciones = tabla_posiciones.sort_values(
    by=['Puntos', 'DG', 'GF'], 
    ascending=[False, False, False]
).reset_index(drop=True)

print(tabla_posiciones)
```

# Tabla de goleadoras mundial 2023

```python
matches_2023 = matches[matches['Year'] == 2023].copy()
```

```python
print("Ejemplos con goles:")
print(matches_2023[matches_2023['home_goal_long'].notna()][
    ['home_team', 'home_goal', 'home_goal_long', 'home_penalty_goal']
].head(5))
```

```python
caso_sweden = matches_2023[
    (matches_2023['home_team'] == 'Sweden') & 
    (matches_2023['home_penalty_goal'].notna())
].iloc[0]

print("Caso Sweden con penal:")
print(f"\nPenalty goal: {caso_sweden['home_penalty_goal']}")
print(f"\nGoal long completo:\n{caso_sweden['home_goal_long']}")
```

```python
def extraer_goleadoras(valor):
    if pd.isna(valor) or valor == '':
        return []
    
    goleadoras = []
    goles = valor.split("', '")
    
    for gol in goles:
        gol = gol.replace("['", "").replace("']", "")
        partes = gol.split('|')
        if len(partes) >= 3:
            nombre = partes[2].strip()
            goleadoras.append(nombre)
    
    return goleadoras
```

```python
def extraer_goleadoras_penalty(valor):
    if pd.isna(valor) or valor == '':
        return []
    
    goleadoras = []
    penales = valor.split('|')
    
    for penal in penales:
        if '(P)' in penal:
            nombre = penal.split('(P)')[0].strip()
            goleadoras.append(nombre)
    
    return goleadoras
```

```python
todas_goleadoras = []

for idx, partido in matches_2023.iterrows():
    #normales
    todas_goleadoras.extend(extraer_goleadoras(partido['home_goal_long']))
    todas_goleadoras.extend(extraer_goleadoras(partido['away_goal_long']))
    
    #de penal
    todas_goleadoras.extend(extraer_goleadoras_penalty(partido['home_penalty_goal']))
    todas_goleadoras.extend(extraer_goleadoras_penalty(partido['away_penalty_goal']))

df_goles = pd.DataFrame(todas_goleadoras, columns=['Jugadora'])

tabla_goleadoras = df_goles.groupby('Jugadora').size().reset_index(name='Goles')

tabla_goleadoras = tabla_goleadoras.sort_values('Goles', ascending=False).reset_index(drop=True)
```

# Tabla completa por equipo y año 

```python
resultados = []

años = sorted(matches['Year'].unique())

for año in años:
    matches_año = matches[matches['Year'] == año].copy()

    host = matches_año['Host'].iloc[0]
    
    home_teams = matches_año['home_team'].unique()
    away_teams = matches_año['away_team'].unique()
    equipos = sorted(list(set(home_teams) | set(away_teams)))

    for equipo in equipos:

        #partidos juados
        partidos_home = matches_año[matches_año['home_team'] == equipo]
        partidos_away = matches_año[matches_año['away_team'] == equipo]

        pj = len(partidos_home) + len(partidos_away)
        
        #goles hechos
        goles_marcados_home = partidos_home['home_score'].sum()
        goles_marcados_away = partidos_away['away_score'].sum()
        goles_totales_marcados = goles_marcados_home + goles_marcados_away
        promedio_goles_marcados = goles_totales_marcados / pj if pj > 0 else 0
        
        #goles recibidos
        goles_recibidos_home = partidos_home['away_score'].sum()
        goles_recibidos_away = partidos_away['home_score'].sum()
        goles_totales_recibidos = goles_recibidos_home + goles_recibidos_away
        promedio_goles_recibidos = goles_totales_recibidos / pj if pj > 0 else 0
        
        #victorias
        victorias_home = (partidos_home['home_score'] > partidos_home['away_score']).sum()
        victorias_away = (partidos_away['away_score'] > partidos_away['home_score']).sum()
        partidos_ganados = victorias_home + victorias_away
        
        #derrotas
        derrotas_home = (partidos_home['home_score'] < partidos_home['away_score']).sum()
        derrotas_away = (partidos_away['away_score'] < partidos_away['home_score']).sum()
        partidos_perdidos = derrotas_home + derrotas_away
        
        #empates
        empates_home = (partidos_home['home_score'] == partidos_home['away_score']).sum()
        empates_away = (partidos_away['away_score'] == partidos_away['home_score']).sum()
        partidos_empatados = empates_home + empates_away
        
        #asistencias
        asistencia_home = partidos_home['Attendance'].sum()
        asistencia_away = partidos_away['Attendance'].sum()
        asistencia_total = asistencia_home + asistencia_away
        promedio_asistencia = asistencia_total / pj if pj > 0 else 0
        
        resultados.append({
            'Año': año,
            'Host': host,
            'Equipo': equipo,
            'Partidos Jugados': pj,
            'Goles Totales Marcados': int(goles_totales_marcados),
            'Promedio Goles Marcados': round(promedio_goles_marcados, 2),
            'Goles Totales Recibidos': int(goles_totales_recibidos),
            'Promedio Goles Recibidos': round(promedio_goles_recibidos, 2),
            'Partidos Ganados': int(partidos_ganados),
            'Partidos Perdidos': int(partidos_perdidos),
            'Partidos Empatados': int(partidos_empatados),
            'Promedio Asistencia': round(promedio_asistencia, 2)
        })

tabla_completa = pd.DataFrame(resultados)
```

### Preguntas clave 

• ¿Cómo ha cambiado el promedio de goles por partido a lo largo de los torneos?

```python
tendencia_goles = matches.groupby('Year').agg({
    'home_score': 'sum',
    'away_score': 'sum',
    'home_team': 'count' 
}).rename(columns={'home_team': 'Total_Partidos'})

tendencia_goles['Goles_Totales'] = tendencia_goles['home_score'] + tendencia_goles['away_score']
tendencia_goles['Promedio_Gol_Partido'] = tendencia_goles['Goles_Totales'] / tendencia_goles['Total_Partidos']

print(tendencia_goles[['Promedio_Gol_Partido']])
```

•¿Existen tendencias en los equipos dominantes y con peor desempeño?

```python
tendencias = tabla_completa.groupby('Equipo').agg({
    'Partidos Jugados': 'sum',
    'Goles Totales Marcados': 'sum',
    'Goles Totales Recibidos': 'sum'
}).reset_index()

tendencias['Diferencia_Gol'] = tendencias['Goles Totales Marcados'] - tendencias['Goles Totales Recibidos']

print("--- Equipos Dominantes (Mejor Diferencia de Gol) ---")
print(tendencias.sort_values('Diferencia_Gol', ascending=False).head(3))

print("\n--- Equipos con dificultades (Peor Diferencia de Gol) ---")
print(tendencias.sort_values('Diferencia_Gol', ascending=True).head(3))
```
